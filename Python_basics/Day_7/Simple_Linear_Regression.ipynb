## Simple Liner Regression

```python
# Step 1 : import library
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
#read data from CSV file
data = pd.read_csv('Salary_dataset.csv')
data.head()
```

```python
#create dataframe of csv file data
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# check any duplicate values are there and store it in a variable
duplicate_rows_df=df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
print(df.isnull().sum())
```

#### Using X to find Y

```python
x=df['YearsExperience'].values
y=df['Salary'].values
y
```

```python
#drop unnamed column
df=df.drop(['Unnamed: 0'],axis=1)   
df.head()
```

#### Training the model

```python
#step 1: train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=47)
```

```python
x_train=x_train.reshape(-1,1)
x_test=x_test.reshape(-1,1)
y_train=y_train.reshape(-1,1)
y_test=y_test.reshape(-1,1)
```

```python
#check shape of train and test data
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#train the model
model.fit(x_train,y_train)  #reshaping because we have only one feature
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### Comparing actual values and predicting values

```python
#predict the model
y_pred=model.predict(x_test)
```

```python
# scatter plot for actual and predicted values
plt.scatter(y_test,y_pred)
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.title("actual vs predicted values")
plt.show()
```

#### Using R2 score to find accuracy of model

```python
from sklearn.metrics import r2_score
r2score=r2_score(y_test,y_pred)
print("Accuracy of model is: ",r2score*100)
print("Mean absolute error: ",mean_absolute_error(y_test,y_pred))
```
